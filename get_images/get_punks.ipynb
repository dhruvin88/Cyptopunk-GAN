```python
from bs4 import BeautifulSoup
import requests
import re
import os
import cloudscraper
import urllib.request
from retrying import retry
from joblib import Parallel, delayed
from tqdm import tqdm
```

```python
scraper = cloudscraper.CloudScraper()
```

```python
if not os.path.exists('../punks'):
    os.mkdir('../punks')
```

```python
@retry(wait_exponential_multiplier=1000, wait_exponential_max=10000,stop_max_attempt_number=10)
def get_html(url):
    html = scraper.get(url).text
    if 'Access denied' in html:
        raise IOError("")
    return html
```

```python
def get_image(url):
    soup = BeautifulSoup(get_html(url))
    images = []
    for img in soup.findAll('img',attrs={'class':'Image--image'}):
        images.append(img.get('src'))
    return images[0]
```

```python
@retry(wait_exponential_multiplier=1000, wait_exponential_max=10000,stop_max_attempt_number=10)
def download_image(url, filename):
    urllib.request.urlretrieve(url, f"../punks/{filename}.jpg")
```

```python
def get_punk(index):
    url = get_image(f"https://opensea.io/assets/ethereum/0xb47e3cd837ddf8e4c57f05d70ab865de6e193bbb/{index}")
    download_image(url, index)
```

```python
Parallel(n_jobs=-1)(delayed(get_punk)(i) for i in tqdm(range(0,10000)))
```
